```python
# Python and Pandas Assessment Notebook

# Import required libraries
import pandas as pd
import numpy as np
from functools import reduce

# Task 1: Create A Dataframe

try:
    df = pd.DataFrame({
        'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 28],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']
    })

    print("Task 1: Original DataFrame")
    print(df)

except Exception as error:
    print("Error creating DataFrame:", error)

print("\n")
```

```python
# Task 2: Row and Column Manipulation

try:
    df_without_city = df.drop(columns=['City'])

    print("Task 2: DataFrame after dropping 'City'")
    print(df_without_city)

except KeyError:
    print("Error: 'City' column not found.")

except Exception as error:
    print("Unexpected error:", error)

print("\n")
```

```python
# Task 3: Handling Null Values
# Creating DataFrame with null values

try:
    df_null = pd.DataFrame({
        'Name': ['Alice', None, 'Charlie'],
        'Age': [25, np.nan, 30],
        'Score': [None, 90, 85]
    })

    print("Task 3: DataFrame with Null Values")
    print(df_null)

    # Fill missing values
    df_filled = df_null.fillna({
        'Name': 'Unknown',
        'Age': 0,
        'Score': 0
    })

    print("\nFilled Null Values:")
    print(df_filled)

    # Alternative method: drop null rows
    print("\nDropping rows with null values:")
    print(df_null.dropna())

except Exception as error:
    print("Error handling null values:", error)

print("\n")
```

```python
# Task 4: GroupBy and Describe

try:
    df_group = pd.DataFrame({
        'Category': ['A', 'B', 'A', 'B', 'A', 'C'],
        'Value': [10, 20, 15, 25, 30, 35]
    })

    grouped_stats = df_group.groupby('Category')['Value'].describe()

    print("Task 4: GroupBy and Describe")
    print(grouped_stats)

    print("\nInterpretation:")
    print("This output shows the count, average, minimum, maximum,")
    print("and quartiles of values for each category.")

except Exception as error:
    print("Error in GroupBy:", error)

print("\n")
```

```python
# Task 5: Concatenation and Merging

try:
    df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
    df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
    df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

    # Vertical concatenation
    concatenated_df = pd.concat([df1, df2], ignore_index=True)

    print("Task 5: Concatenated DataFrame")
    print(concatenated_df)

    # Horizontal merge
    merged_df = pd.concat([df1, df3], axis=1)

    print("\nMerged DataFrame")
    print(merged_df)

except Exception as error:
    print("Error in concatenation or merging:", error)

print("\n")
```

```python
# Task 6: Tuples and Sets

try:
    fruits_tuple = ('apple', 'banana', 'orange')
    numbers_set = {1, 2, 3, 4, 5}

    print("Task 6: Tuple and Set")
    print("Original tuple:", fruits_tuple)
    print("Original set:", numbers_set)

    # Attempt to modify tuple
    try:
        fruits_tuple.append('grape')
    except AttributeError as error:
        print("Tuple Error:", error)

    # Add to set
    numbers_set.add(6)

    print("Updated set:", numbers_set)

    print("\nExplanation:")
    print("Tuples cannot be changed after creation (immutable).")
    print("Sets can be changed (mutable).")

except Exception as error:
    print("Error with tuple/set:", error)

print("\n")
```

```python
# Task 7: Dictionaries

try:
    student_scores = {
        'Alice': 85,
        'Bob': 90,
        'Charlie': 78
    }

    # Update existing student
    if 'Bob' in student_scores:
        student_scores['Bob'] = 95

    # Add new student
    student_scores['David'] = 88

    print("Task 7: Dictionary")
    print(student_scores)

except Exception as error:
    print("Dictionary error:", error)

print("\n")
```

```python
# Task 8: Functions and Lambda

def square_number(number):
    try:
        return number ** 2
    except TypeError:
        return "Invalid input"

square_lambda = lambda number: number ** 2 if isinstance(number, (int, float)) else "Invalid input"

print("Task 8: Functions and Lambda")
print("Regular function (4):", square_number(4))
print("Regular function (7):", square_number(7))

print("Lambda function (4):", square_lambda(4))
print("Lambda function (7):", square_lambda(7))

print("\n")
```

```python
# Task 9: Iterators and Generators

class EvenNumberIterator:
    def __init__(self):
        self.current = 0
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < 5:
            even_number = self.current
            self.current += 2
            self.count += 1
            return even_number
        else:
            raise StopIteration


def even_number_generator():
    for number in range(0, 10, 2):
        yield number


print("Task 9: Iterator")
try:
    for number in EvenNumberIterator():
        print(number)
except Exception as error:
    print("Iterator error:", error)

print("\nTask 9: Generator")
try:
    for number in even_number_generator():
        print(number)
except Exception as error:
    print("Generator error:", error)

print("\n")
```

```python
# Task 10: Map, Reduce, and Filter

numbers_list = [1, 2, 3, 4, 5]

try:
    squared_numbers = list(map(lambda x: x ** 2, numbers_list))

    if numbers_list:
        product_numbers = reduce(lambda x, y: x * y, numbers_list)
    else:
        product_numbers = "List is empty"

    even_numbers = list(filter(lambda x: x % 2 == 0, numbers_list))

    print("Task 10: Map, Reduce, Filter")
    print("Squared numbers:", squared_numbers)
    print("Product of numbers:", product_numbers)
    print("Even numbers:", even_numbers)

except Exception as error:
    print("Error in map/reduce/filter:", error)

print("\n")
```

```python
# Task 11: Object-Oriented Programming

class Rectangle:
    def __init__(self, length, width):
        if length <= 0 or width <= 0:
            raise ValueError("Length and width must be positive values.")
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)


try:
    rectangle1 = Rectangle(5, 3)
    rectangle2 = Rectangle(10, 4)

    print("Task 11: Rectangle Class")

    print("Rectangle 1 Area:", rectangle1.calculate_area())
    print("Rectangle 1 Perimeter:", rectangle1.calculate_perimeter())

    print("Rectangle 2 Area:", rectangle2.calculate_area())
    print("Rectangle 2 Perimeter:", rectangle2.calculate_perimeter())

except ValueError as error:
    print("Rectangle error:", error)

print("\n")
```
