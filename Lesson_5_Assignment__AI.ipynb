```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Task 1: Load the Titanic dataset

df = pd.read_csv('/content/Titanic-Dataset.csv')

# Display dataset loaded successfully
df.head()
```

```python
# Task 2: Inspect the data

# First 5 rows
print("First 5 rows:")
print(df.head())

# Last 5 rows
print("Last 5 rows:")
print(df.tail())

# Data types
print("Data types:")
print(df.dtypes)

# Summary statistics
print("Summary statistics:")
print(df.describe())
```

```python
# Task 3: Clean and preprocess the data

# Create Pclass category
df['PclassCategory'] = df['Pclass'].astype('category')

# Create Cabin category
df['CabinCategory'] = df['Cabin'].fillna('Unknown').str[0]

# Handle missing Age values (fixed warning)
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Create Family feature
df['Family'] = df['SibSp'] + df['Parch']

# Show only a small preview
print(df[['PclassCategory', 'CabinCategory', 'Age', 'Family']].head(3))

# Check missing values
print("\nMissing Age values:", df['Age'].isnull().sum())
```

```python
# Task 4: Identify specific data points

#vHighest fare passengers

highest_fare_passengers = df[df['Fare'] == df['Fare'].max()]

print("Passenger(s) with highest fare:")
print(highest_fare_passengers[['Name', 'Fare']])

# Largest family size

largest_family_passengers = df[df['Family'] == df['Family'].max()]

print("Passenger(s) with largest family size:")
print(largest_family_passengers[['Name', 'Family']])
```

```python
# Task 5: Pandas Series and DataFrame

# Series of names

names_series = df['Name']

print("Passenger names series:")
print(names_series.head())

# DataFrame of names and ages

names_ages_df = df[['Name', 'Age']]

print(names_ages_df.head())

# Passengers older than 50
older_passengers = df[df['Age'] > 50]

print("Passengers older than 50:")
print(older_passengers[['Name', 'Age']].head())
```

```python
# Task 6: Multi-indexing

multi_index_df = df.groupby(['Sex', 'Pclass'])['Survived'].mean()

print("Survival rates by Sex and Pclass:")
print(multi_index_df)
```

```python
# Task 7: Visualizations

# Histogram of passenger ages

plt.figure(figsize=(4,3))
plt.hist(df['Age'], bins=30)
plt.title('Histogram of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Insight: Most passengers were between 20 and 40 years old. Very few passengers were elderly.

```python
# Task 7: Visualizations

# Bar plot of survival rates by class

plt.figure(figsize=(4,3))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rates by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

Insight: Passengers in first class had the highest survival rate.

```python
# Task 7: Visualizations

# Heatmap of correlations

plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

Insight: Fare and Pclass show strong correlation. Survival is slightly linked to Fare and Pclass.

```python
# Task 7: Visualizations

# Pair plot

sns.pairplot(df[['Age', 'Fare', 'SibSp', 'Parch', 'Survived']], hue='Survived')
plt.show()
```

Insight: Passengers who paid higher fares had a better chance of surviving.

```python
# Task 8: Analyze survival rates

# By sex

survival_by_sex = df.groupby('Sex')['Survived'].mean()

print("Survival by sex:")
print(survival_by_sex)

# By passenger class

survival_by_class = df.groupby('Pclass')['Survived'].mean()

print("Survival by class:")
print(survival_by_class)
```

Insight: Females survived more often than males. First-class passengers survived more than lower classes.

```python
# Task 9: Box plot (Age distribution)

plt.figure(figsize=(4,3))
sns.boxplot(x='Pclass', y='Age', hue='Sex', data=df)
plt.title('Age Distribution by Passenger Class and Sex')
plt.show()
```

Insight: First-class passengers tended to be older on average.

```python
# Task 10: Violin plot (Fare distribution)

plt.figure(figsize=(4,3))
sns.violinplot(x='Pclass', y='Fare', hue='Survived', data=df)
plt.title('Fare Distribution by Passenger Class and Survival Status')
plt.show()
```

Insight: Higher ticket fares were often linked with survival.
